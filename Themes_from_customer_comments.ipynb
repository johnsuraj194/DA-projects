```python
# import nltk
# nltk.download('averaged_perceptron_tagger')

def join_consecutive_words_with_pos(words):
    pos_tags = nltk.pos_tag(words)
    joined_words = []
    prev_tag = None
    prev_prev_tag = None
    for i in range(len(pos_tags)):
        word, tag = pos_tags[i]
        if prev_prev_tag == 'JJ' and prev_tag == 'NN' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_prev_tag == 'JJ' and prev_tag == 'NNS' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_prev_tag == 'JJ' and prev_tag == 'NN' and tag == 'NNS':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_prev_tag == 'NN' and prev_tag == 'NN' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_prev_tag == 'NN' and prev_tag == 'VB' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        
        elif prev_prev_tag == 'VBG' and prev_tag == 'JJ' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_prev_tag == 'VBG' and prev_tag == 'JJ' and tag == 'NNS':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_prev_tag == 'VBG' and prev_tag == 'NN' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the three consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
            
        elif prev_tag == 'NN' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        
        elif prev_tag == 'NN' and tag == 'NNS':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_tag == 'NNS' and tag == 'JJ':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_tag == 'JJ' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_tag == 'JJ' and tag == 'JJ':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_tag == 'JJ' and tag == 'NNS':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_tag == 'VB' and tag == 'NN':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        elif prev_tag == 'VB' and tag == 'NNS':
            if len(joined_words) > 0 and joined_words[-1][-1] != ' ':
                # add space between the two consecutive nouns
                joined_words[-1] += ' '
            joined_words[-1] += word
        else:
            joined_words.append(word)
        prev_prev_tag = prev_tag
        prev_tag = tag
    return joined_words
```
